```
from statsmodels.stats.multitest import multipletests
from statannotations.Annotator import Annotator
from scipy import stats
import sys

def multi_class(data, cn, fnames, method):
    if len(data.groupby(cn)) <= 2:
        raise Exception('ANOVA requires a secondary index with three or more values')
    if method=='anova':
        func = stats.f_oneway
    elif method=='kruskal':
        func = stats.kruskal
    vstats = []
    for col in fnames:
        cls = []
        for k, v in data[[cn,col]].groupby(cn):
            cls.append(v[col])
        try:
            s = func(*cls)
        except:
            s = (0,1)
        vstats.append(s)
    return pd.DataFrame(vstats)
```

```
len(set(grp))
```

```
# Teste mais robusto, menos sensivel
tkruskal = multi_class(res, 'group', X_norm.columns, 'kruskal')
tkruskal.head()
```

```
sum(tkruskal['pvalue'] < 0.05)
```

```
tkruskal['pval_cor'] = multipletests(tkruskal['pvalue'], method='fdr_bh')[1]
sum(tkruskal['pval_cor'] < 0.05)
```

```
# pressuposicoes para teste parametrico razoaveis para dados de metabolomica
tanova = multi_class(res, 'group', X_norm.columns, 'anova')
tanova['pval_cor'] = multipletests(tanova['pvalue'], method='fdr_bh')[1]
sum(tanova['pval_cor'] < 0.05)
```

```
import itertools
aflavus = [x for x in grp if 'Aflavus' in x]
pairs = list(itertools.combinations(set(aflavus), 2))
pairs
```

```
plotting_parameters = {
    'data': res[res.group.str.contains('Aflavus')],
      'x': 'group',
      'y': 6
}

#tmp[['classe_mortalidade', i]].groupby('classe_mortalidade').mean()
pvalues = []
for p in pairs:
    pvalues.append(stats.mannwhitneyu(res.loc[res.group==p[0], 9],
                                 res.loc[res.group==p[1], 9]).pvalue)

formatted_pvalues = [f"p={p:.2e}" for p in pvalues]
#ax = get_log_ax()
fig, ax = plt.subplots(1, 1, figsize=(12,6)) 
sns.boxplot(**plotting_parameters)
sns.stripplot(**plotting_parameters)
annotator = Annotator(ax, pairs, **plotting_parameters)
annotator.set_pvalues(pvalues)
annotator.annotate()
plt.xticks(rotation=45)
```
